```python

import pandas as pd

job=pd.read_excel('CompiledAnalyticsData_excel.xlsx')
job.head(3)
```

```python
#job['skill']=job['skills'].replace('[^A-Za-z0-9]+',' ',regex=True)
```

```python
job['JobClass']=job['Job Class']
job['CleanJobDescription']=job['Clean Job Description']
```

```python
cleanjb=job['CleanJobDescription'].values.astype('str')
```

```python
jobgroup = job.groupby("JobClass")
len(jobgroup)
jobgroup.head()
```

```python
jobgroup.describe()
```

```python
print(job.shape)
```

```python

#split the data into training and testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(cleanjb, job.JobClass, test_size=0.33, random_state=12)

```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
```

```python
#Build training pipeline using  Naïve Bayes Model

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

text_clf = Pipeline([('vect', CountVectorizer()),   #Vectorizer
                     ('tfidf', TfidfTransformer()), #DTM with TFIDF
                      ('clf', MultinomialNB()),     #ML Model
                    ])
```

```python
text_clf.fit(X_train,y_train) 
```

```python
#Evaluate the model

import numpy as np
from sklearn import metrics

predicted = text_clf.predict(X_test)

print(metrics.confusion_matrix(y_test, predicted))
```

```python
print("NB:",np.mean(predicted == y_test) )
```

```python
print(metrics.classification_report(y_test, predicted))
```

```python
#Build training pipeline using  Decision Tree Model


from sklearn import tree
text_clf = Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf', tree.DecisionTreeClassifier())])
clf = text_clf.fit(X_train, y_train) 

predicted = clf.predict(X_test)
  
```

```python
print(metrics.confusion_matrix(y_test, predicted))
```

```python
print("DT:",np.mean(predicted == y_test) )
```

```python
print(metrics.classification_report(y_test, predicted))
```

```python
#Build training pipeline using  SVM Model


from sklearn import svm
from sklearn.svm import SVC

from sklearn.linear_model import SGDClassifier
text_clf = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer(use_idf=True)),
                      ('clf', svm.LinearSVC(C=1.0))
                    ])
text_clf.fit(X_train, y_train) 
    
predicted = text_clf.predict(X_test)
 
```

```python
print(metrics.confusion_matrix(y_test, predicted))
```

```python
print(np.mean(predicted == y_test) )
```

```python
print(metrics.classification_report(y_test, predicted))
```
