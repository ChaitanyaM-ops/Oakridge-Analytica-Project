```python
import numpy as np
import pandas as pd

job=pd.read_excel('CompiledAnalyticsData_excel.xlsx')
job.head(3)
```

```python
job['JobClass']=job['Job Class']
job['JobTitle']=job['Clean Job Title']
```

```python
jobgroup = job.groupby("JobClass")
len(jobgroup)
jobgroup.head()
```

```python
jobgroup.describe()
```

```python
print(job.shape)
```

```python

#split the data into training and testing

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(job.JobTitle, job.JobClass, test_size=0.33, random_state=12)

```

```python
from sklearn.feature_extraction.text import CountVectorizer
count_vect = CountVectorizer()
```

```python
#Build training pipeline using  Naïve Bayes Model

from sklearn.pipeline import Pipeline
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.naive_bayes import MultinomialNB

text_clf1 = Pipeline([('vect', CountVectorizer()),   #Vectorizer
                     ('tfidf', TfidfTransformer()), #DTM with TFIDF
                      ('clf', MultinomialNB()),     #ML Model
                    ])

clf1=text_clf1.fit(X_train,y_train ) 
```

```python
#Evaluate the model

import numpy as np
from sklearn import metrics

predicted1 = text_clf1.predict(X_test)

print(metrics.confusion_matrix(y_test, predicted1))
```

```python
print("NB:",np.mean(predicted1 == y_test) )
```

```python
print(metrics.classification_report(y_test, predicted1))
```

```python
#Build training pipeline using  Decision Tree Model


from sklearn import tree
text_clf2 = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer()),
                      ('clf', tree.DecisionTreeClassifier())
                    ])
clf2 = text_clf2.fit(X_train, y_train) 

predicted2 = clf2.predict(X_test)
  
```

```python
print(metrics.confusion_matrix(y_test, predicted2))
```

```python
print("DT:",np.mean(predicted2 == y_test) )
```

```python
print(metrics.classification_report(y_test, predicted2))
```

```python
#Build training pipeline using  SVM Model


from sklearn import svm
from sklearn.svm import SVC

from sklearn.linear_model import SGDClassifier
text_clf3 = Pipeline([('vect', CountVectorizer()),
                     ('tfidf', TfidfTransformer(use_idf=True)),
                      ('clf', svm.LinearSVC(C=1.0))
                    ])
text_clf3.fit(X_train, y_train) 
    
predicted3 = text_clf3.predict(X_test)
 
```

```python
print(metrics.confusion_matrix(y_test, predicted3))
```

```python
print(np.mean(predicted3 == y_test) )
```

```python
print(metrics.classification_report(y_test, predicted3))
```
